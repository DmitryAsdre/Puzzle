```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
names = ['test_1', 'test_2', 'test_3', 'test_4']
```

```python
for name in names:
    df = pd.read_csv(name + '.csv')
    print(name)
    plt.figure(figsize=(15, 7))
    mask = df['Ratio'] > 10
    plt.xlabel("Ratio")
    plt.ylabel("#ofOperations")
    plt.plot(df['Ratio'][mask], df['#ofOperations'][mask])
    plt.plot()
```

```python
mask = df['Ratio'] > 1.2
plt.figure(figsize=(15, 7))
plt.plot(df['Ratio'][mask], df['#ofOperations'][mask])
```
